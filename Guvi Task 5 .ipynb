```python
import pandas as pd
```

```python
df1 = pd.read_csv('takehome_user_engagement.csv')
df1.head()
```

```python
df2 = pd.read_csv('takehome_users.csv',encoding ='latin-1')
df2.head()
```

```python
df1.shape
```

```python
df2.shape
```

```python
df2.rename(columns = {'object_id' : 'user_id'}, inplace = True)
df2.head()
```

```python
df = pd.merge(df2,df1, on = ['user_id'], how = 'left')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.drop_duplicates()
```

```python
df["time_stamp"] = pd.to_datetime(df["time_stamp"])
```

```python
df['day'] = df["time_stamp"].dt.date
```

```python
df['month'] = df["time_stamp"].dt.month
```

```python
df['year'] = df["time_stamp"].dt.isocalendar().year
```

```python
df['week'] = df["time_stamp"].dt.isocalendar().week
```

```python
df.head()
```

```python
week_data = df.groupby(['user_id','name','week'])['visited'].count().reset_index()
week_data
```

```python
week_data[week_data['visited'] >=3]
```

```python
week_data['adopted_user'] = week_data['visited'].apply(lambda x : 1 if x>= 3 else 0)
```
